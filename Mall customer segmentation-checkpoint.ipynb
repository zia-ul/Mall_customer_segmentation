```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
```

```python
#load data 
df = pd.read_csv('Mall_Customers.csv')
```

```python
df.head()
```

```python
df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(10, 6))
sns.set(style = 'whitegrid')
sns.distplot(df['Annual Income (k$)'])
plt.title('Distribution of Annual Income (k$)', fontsize = 20)
plt.xlabel('Range of Annual Income (k$)')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(10, 6))
sns.set(style = 'whitegrid')
sns.distplot(df['Spending Score (1-100)'])
plt.title('Distribution of Spending Score (1-100)', fontsize = 20)
plt.xlabel('Range of Spending Score (1-100)')
plt.ylabel('Count')
```

```python
Genders = df["Gender"].value_counts()

colors = {
    "Male": "blue",
    "Female": "orange"
}

sns.set(style="whitegrid")
sns.barplot(x=Genders.index, y=Genders.values, palette=colors)

plt.show()
```

```python
X = df[["Annual Income (k$)", "Spending Score (1-100)"]]
X.head()
```

```python
sns.scatterplot(x="Annual Income (k$)", y="Spending Score (1-100)", data=X, s=60, color="blue")

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Spending Score (1-100) vs Annual Income (k$)')

plt.show()
```

```python
# Determine the optimal number of clusters (k) using the Elbow method
wcss = []
for i in range(1, 11):
    k_means = KMeans(n_clusters=i, n_init="auto")
    k_means.fit(X)
    wcss.append(k_means.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.plot(range(1, 11), wcss, linewidth=2.5, color="red", marker="8")
# Plot the Elbow graph to choose the optimal k
plt.title("Elbow Curve")
plt.xlabel("Number Of Clusters (values of k)")
plt.ylabel("WCSS values")
plt.xticks(np.arange(1, 11, 1))

plt.show()
```

```python
k_means_1 = KMeans(n_clusters=3, n_init="auto")   # K = 3 CLUSTERS
k_means_1.fit(X)
y_pred = k_means_1.predict(X)
df["label"] = y_pred
df.head()
```

```python
colors = ["orange", "indigo", "black"]

sns.scatterplot(
    x="Annual Income (k$)",
    y="Spending Score (1-100)",
    hue="label", 
    palette=colors, 
    legend="full", 
    data=df, 
    s=60
)

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)') 
plt.title('Spending Score (1-100) vs Annual Income (k$)')
plt.show()
```

```python
df2=df[["CustomerID","Gender","Age","Annual Income (k$)","Spending Score (1-100)"]]
X2=df2[["Age","Annual Income (k$)","Spending Score (1-100)"]]
```

```python
# Determine the optimal number of clusters (k) using the Elbow method
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X2)
    wcss.append(kmeans.inertia_)
# Plot the Elbow graph to choose the optimal k
plt.figure(figsize=(12,6))    
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

```python
km2 = KMeans(n_clusters=5)
y2 = k_means.fit_predict(X2)
df2["label"] = y2

df2.head()
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df2.Age[df2.label == 0], df2["Annual Income (k$)"][df2.label == 0], df2["Spending Score (1-100)"][df2.label == 0], c='orange', s=60)
ax.scatter(df2.Age[df2.label == 1], df2["Annual Income (k$)"][df2.label == 1], df2["Spending Score (1-100)"][df2.label == 1], c='indigo', s=60)
ax.scatter(df2.Age[df2.label == 2], df2["Annual Income (k$)"][df2.label == 2], df2["Spending Score (1-100)"][df2.label == 2], c='black', s=60)
ax.view_init(35, 185)
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```
